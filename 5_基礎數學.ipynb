## ÂáΩÊï∏ $f(x) = x^2 + 3x -5$
## Â∞éÂáΩÊï∏ $f'(x) = 2x + 3$
- ËóçËâ≤ÂàáÁ∑öÈÄöÈÅé $(x,y)=(1,-1)$ ÁöÑÂàáÈªûÔºåÁï∂ $x=1$ ÁöÑÈªûÈÄêÊº∏ÂæÄ $x=2$ Èù†ËøëÔºåÈÄêÊº∏Èù†ËøëÂà∞ÂÖ©ÂÄãÈªûÁöÑÂ∑ÆË∑ù $\Delta x$ ÈÄºËøëÊñº 0Ôºå‰æãÂ¶ÇÂ∫ï‰∏ãÁöÑÂúñÁ§∫„ÄÇ
- Êï∏Â≠∏ÂºèÁöÑË°®ÈÅîÊñπÂºè $f'(x) = lim_{x\rightarrow 2}\frac{f(x)-f(2)}{x-2}$ ÔºåÊÑèÂë≥Ëëó $f(x)$ Âú® 2 ÁöÑÂàáÈªûÁöÑÊñúÁéáÁ≠âÊñº $f(x)$ Âú® 2 Ê≠§ÈªûÁöÑÈô°Â≥≠Á®ãÂ∫¶„ÄÇ
- ÊâÄ‰ª•ÔºåÊ≥õÂåñÂá∫‰∏ÄÂÄãÂ∞éÂáΩÊï∏ $f'(x) = lim_{h\rightarrow 0}\frac{f(x+h)-f(x)}{h}$ÔºåÂè™Ë¶ÅÂ∏∂ÂÖ•ÊÉ≥Ë¶ÅÁöÑÈªûÔºå‰æøÂèØÊ±ÇÂá∫Ë©≤ÈªûÂú®ÂáΩÊï∏‰∏äÁöÑÂàáÁ∑öÊñúÁéáÔºåÂÖ∂Ê≠£Ë≤†Ëôü‰πüË°®Á§∫Êï¥ÂÄãÂáΩÊï∏ÁöÑ‰∏äÂçáÊàñ‰∏ãÈôçÊñπÂêë„ÄÇ

## ÂêàÊàêÂáΩÊï∏
- ÂêàÊàêÂáΩÊï∏ÂèØÂ∞áË§áÈõúÁöÑÂºèÂ≠êÁ∞°Âåñ„ÄÇ
- ÂêàÊàêÂáΩÊï∏ÁöÑÊ¶ÇÂøµÊòØÁî®Á∞°ÂñÆÁöÑÂáΩÊï∏ÂêçÁ®±ÔºåÂèñ‰ª£Áúã‰ººË§áÈõúÁöÑÂáΩÊï∏„ÄÇ

### ÂÅáË®≠ $y = f(x) = -100(x - 2.8)^2 + 1000$

### ‰ª§ $ u = x - 2.8$
### $y = f(x) = -100\boxed{(x - 2.8)}^2 + 1000$
### $\rightarrow f(u) = -100u^2 + 1000$

### ‰ª§ $ u = g(x) = x - 2.8$
### ÊâÄ‰ª• $y = f(u) = -100u^2 + 1000$„ÄÇ
### Â∞á $f$ Ëàá $g$ ÂÖ©ÂÄãÂáΩÊï∏ÂêàÊàêÂú®‰∏ÄËµ∑Á≠âÊñº $y$
### $\rightarrow \boxed{y = f(g(x))}$

## ÈÄ£ÈéñÊ≥ïÂâá (Chain Rule)

### Áï∂ $y$ Â∞ç $x$ ÂæÆÂàÜÔºå‰æãÂ¶Ç‰∏ä‰æãÁöÑÂêàÊàêÂáΩÊï∏ $f(g(x))$ÔºåÂøÖÈ†àÂàÜÂà•Â∞ç $f$ Ëàá $g$ ÂáΩÊï∏ÈÄ£Á∫åÂæÆÂàÜ:
### $\rightarrow y' = f'(g(x)) \cdot g'(x)$
### $= \boxed{f'(u)\cdot u'}$

### Â¶ÇÊûúÂêàÊàêÂáΩÊï∏ÊòØÁî±Êõ¥Â§öÁöÑÂáΩÊï∏‰∏ÄÂ±§‰∏ÄÂ±§ÊâÄÁµÑÁπîÂêàÊàêÔºå‰πüÂêåÊ®£ÈÅ©Áî®ÈÄ£ÈéñÊ≥ïÂâáÔºå‰æãÂ¶Ç:
### $y = f(g(h(x)))$ÔºåÂ∞á y Â∞ç x ÂæÆÂàÜ
### $\rightarrow \boxed{y' = f'(g(h(x)) \cdot g'(h(x)) \cdot h'(x)}$

## ‰ΩøÁî®Â∏ÉËêäÂ∞ºËå≤Ë®òÊ≥ïË°®Á§∫ÂêàÊàêÂáΩÊï∏ÁöÑÈÄ£ÈéñÊ≥ïÂâáÂæÆÂàÜ
### $\frac{dy}{dx} = \frac{dy}{du}\cdot \frac{du}{dx}$
- $y$ Â∞ç $x$ÂæÆÂàÜÔºåÁ≠âÂêåÊñº„Äå$y$Â∞ç$u$ÂæÆÂàÜ„Äç‰πò‰ª•„Äå$u$Â∞ç$x$ÂæÆÂàÜ„Äç
### Âõ†ÁÇ∫$ y = f(u) = f(g(x)) $
### ÊâÄ‰ª•Ôºå$ \boxed{\frac{dy}{dx} = \frac{dy}{du}\cdot \frac{du}{dx}} \equiv \boxed{f'(u)\cdot u'}$

## ÁØÑ‰æã

## $\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx}$

### ËÄå $\frac{dy}{du} = \frac{d}{du}f(u) = f'(u) = -200\cdot u = -200u$

### ËÄå $\frac{du}{dx} = \frac{d}{dx}(x - 2.8) = 1$

### ÊâÄ‰ª•Ôºå$\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} = -200u \cdot 1 = -200(x - 2.8)$

## ÂÅèÂ∞éÊï∏

## Â§öËÆäÊï∏ÂáΩÊï∏
- ÊúâÂÖ©ÂÄãÊàñÂÖ©ÂÄã‰ª•‰∏äËÆäÊï∏ÁöÑÂáΩÊï∏ÔºåÁ®±ÁÇ∫Â§öËÆäÊï∏ÂáΩÊï∏„ÄÇ

### ÂáΩÊï∏ $z = f(x, y) = x + y$ÔºåÂÖ∂‰∏≠ $x$ Ëàá $y$ ‰∫íÁõ∏Áç®Á´ãÔºå‰∏çÂèóÂΩ±Èüø„ÄÇ

### $\frac{\partial z}{\partial x} = \frac{\partial}{\partial x}(x + y) = \frac{\partial}{\partial x}(x) + \frac{\partial}{\partial x}(y) = 1 + 0 = 1$

- **Ê¢ØÂ∫¶ÔºåÂèØ‰ª•Ë°®ÁèæÊ≠§È´òÁÇ∫Êõ≤Èù¢Âú®Êüê‰∏ÄÈªû‰∏äÔºåÂ∞çÊñºÊØè‰∏ÄÂÄãÊñπÂêëÁöÑÊèèËø∞„ÄÇ‰∏çÂÉÖÂÄãÂà•ÂéªÁúãÊ≠§ÈªûÂú®ÊØè‰∏ÄÂÄãËª∏‰∏äÁöÑÂÇæÊñúË°®ÁèæÔºåÊ¢ØÂ∫¶ÂåÖÂê´‰∫ÜÊâÄÊúâÁöÑË≥áË®äÂú®ÂÖ∂‰∏≠„ÄÇ**

## ÂÜç‰ª•ÂâçËø∞ $f(x,y)=e^{2y}sin(x)$ÂáΩÊï∏ÁÇ∫‰æã
### ÂáΩÊï∏ $f$ ÂÉÖÈáùÂ∞çËÆäÊï∏ $x$ ÂÅèÂæÆÂàÜÔºåÂâá $y$ ËÆäÊï∏Ë¶ñÁÇ∫Â∏∏Êï∏ $\frac{\partial f}{\partial x} (x, y) = \boxed{e^{2y}cos(x)}$
### ÂáΩÊï∏ $f$ ÂÉÖÈáùÂ∞çËÆäÊï∏ $y$ ÂÅèÂæÆÂàÜÔºåÂâá $x$ ËÆäÊï∏Ë¶ñÁÇ∫Â∏∏Êï∏$\frac{\partial f}{\partial y} (x, y) = \boxed{2e^{2y}sin(x)}$

### ÈÇ£È∫ºÔºåÂáΩÊï∏$f(x,y)=e^{2y}sin(x)$ÁöÑÊ¢ØÂ∫¶‰∏ãÈôç$\Delta f(x,y)$Ë°®ÈÅîÂ¶Ç‰∏ã:
### $\triangledown f(x,y) = (\frac{\partial f}{\partial x}(x, y), \frac{\partial f}{\partial y}(x, y)) = \boxed{(e^{2y}cos(x), 2e^{2y}sin(x))}$

## Ê¢ØÂ∫¶‰∏ãÈôç (Gradient Descent)
- „ÄåÊ¢ØÂ∫¶‰∏ãÈôç„ÄçÊºîÁÆóÊ≥ï (Gradient Descent Algorithm)ÊòØÂ§ßÂ§öÊï∏Ê©üÂô®Â≠∏Áøí( machime Learning ÔºâÂíåÊ∑±Â∫¶Â≠∏Áøí (Deep Learning) ÁöÑÊ†∏ÂøÉÂÖßÂÆπ„ÄÇ
- „ÄåÊ¢ØÂ∫¶‰∏ãÈôç„ÄçÊºîÁÆóÊ≥ïÁöÑÁõÆÊ®ôÊòØË¶ÅÁõ¥Âà∞ÊâæÂà∞‰∏ÄÂÄãÂáΩÊï∏ÁöÑÊúÄ‰Ω≥Ëß£ÔºåÊ¶ÇÂøµÊòØÊ†πÊìöÈö®Ê©üÈÅ∏ÊìáÂáΩÊï∏ÁöÑÂàùÂßãÈªûÔºåËóâÁî±Ëø≠‰ª£ (Iterations) Ë®àÁÆóÔºåÔºåÂæ™Â∫èÊº∏ÈÄ≤Âú∞ÊêúÂ∞ãÔºå‰∏çÊñ∑Ë™øÊï¥Ê≠•‰ºê‰ª•ÂèäÊñπÂêëÔºåË©¶ÂúñÊâæÂá∫ÊúÄÈÅ©ÂêàÁöÑËß£„ÄÇ
- Âú®‰∏ÄÂÄãÊêçÂ§±ÂáΩÊï∏ (Loss Function) ‰∏äÔºåÈö®Ê©üÊâæ‰∏ÄÂÄãÈªûÔºåÈÄ≤Ë°åËø≠‰ª£ÔºåÁßªÂãïÂà∞‰∏ã‰∏ÄÂÄãÈªûÔºåÁõ¥Âà∞‰∏çÂÜçÁßªÂãïÊàñÁÑ°Ê≥ïÁßªÂãïÁÇ∫Ê≠¢„ÄÇ

### $$w^{(t+1)} = w^t + \eta \cdot (-\triangledown L(w^t)) = \boxed{w^t - \eta \cdot \triangledown L(w^t)}$$

### Ë™™Êòé : ‰∏ãÈôçÊ¢ØÂ∫¶Êñ∞ÁöÑÈªû ($w^{t+1}$) ÁöÑ‰ΩçÁΩÆÊòØ‰æùÈù†ËàäÁöÑÈªû ($W^t$) Ê¢ØÂ∫¶Ê≠£Ë≤†ËôüÁöÑÂèçÊñπÂêëÔºå‰ª•ÂèäÂ≠∏ÁøíÁéáÂ§ßÂ∞è ($\eta$) Ë®àÁÆóÂæóÁü•„ÄÇ
- $w^t$: ÁõÆÂâçÁöÑ‰ΩçÁΩÆ
- $w^{(t+1)}$: Êñ∞ÁöÑ‰ΩçÁΩÆ
- $t$: iteration Ëø≠‰ª£, $t\leftarrow$ ÊòØÂæûÁ¨¨ 0 Ê¨°ÈñãÂßã<br>
- $\eta$: Â≠∏ÁøíÁéá (Learning Rate)
- $\triangledown L$: Ê¢ØÂ∫¶‰∏ãÈôçÂáΩÊï∏ÁöÑËÆäÂåñ
- $-\triangledown L(w^t)$: Ë®≠ÂÆöË≤†ËôüË°®Á§∫ÊêçÂ§±(loss)ÊÑàÂ∞èÊÑàÂ•ΩÔºåÁ¢∫ÂÆöÊ¢ØÂ∫¶ÊòØÂæÄÊúÄÂ∞èÊñπÂêëÁßªÂãï„ÄÇ

### Ê¢ØÂ∫¶‰∏ãÈôçÊ≥ïÊºîÁÆóÊ≥ï : 
$1: Guess\; w^{(0)}, set \; t\leftarrow \, 0 $<br>
$2: \textbf{while} \,||\triangledown f(w^{(t)}) || \geq \epsilon \; \textbf{do}  $<br>
$3: \;\;\;\; w^{(k+1)} = |w^{(k)} - \eta \triangledown f(w^{(k)})|  $<br>
$4: \;\;\;\; t \leftarrow t + 1 $<br>
$5: \textbf{end while} $<br>
$6: \textbf{return} \;\; w^{(k)}$ <p>

### Ê¢ØÂ∫¶‰∏ãÈôçÊ≥ïÊºîÁÆóÊ≥ïË™™Êòé:
$w^{(0)}: w \;ÁöÑÂàùÂßãË®≠ÂÆö$ <br>
$t: iteration$;  $t\leftarrow \; 0$ ÊòØÂæûÁ¨¨ 0 Ê¨°ÈñãÂßã<br>
$\triangledown f(w^{(t)}):\; $ Ë°®Á§∫Ê¢ØÂ∫¶‰∏ãÈôçÂáΩÊï∏Âú®Á¨¨ t Ê¨°ÁöÑËÆäÂåñÁéá <br>
$||\triangledown f(w^{(t)}) || \geq \epsilon:\;$ Ë°®Á§∫ÊòØÁ¨¨ t ÂÄã iteration ÊôÇÔºåËã•Â§ßÊñº $\epsilon$ ËÆäÂåñÂÄºÔºåÂâáÈÄ≤ÂÖ• iteration <br>
$ w^{(k+1)} = w^{(k)} - \eta \triangledown f(w^{(t)}): \;$ <br>
$ \;\;\;\; w^{(t+1)}\; ÊòØÊñ∞ÁöÑ w$ <br>
$ \;\;\;\; w^{(t)}\; ÊòØËàäÁöÑ x, $ <br>
$ \;\;\;\; \eta \,ÊòØ\;learning \;rate$ <br>
$\therefore Êñ∞ÁöÑ w^{(t+1)} \; Á≠âÊñºËàäÁöÑ \;w^k\;Ê∏õÂéª\;learning rate \;\eta \;‰πò‰ª•Ê¢ØÂ∫¶‰∏ãÈôç\; \triangledown f(w^{(t)})$ <br>
$t \leftarrow t + 1: \;$ ÂÜçÈÄ≤ÂÖ•‰∏ã‰∏ÄÂÄãÂæ™Áí∞<br>

## Ê¢ØÂ∫¶‰∏ãÈôç (Gradient Descent) ÁâπÊÄß
- ÂøÖÈ†àË¶ÅÁ¢∫ÂÆö Loss function Êú¨Ë∫´ÂèØÂæÆÂàÜ (differentiable)„ÄÇ

- Â≠∏ÁøíÁéá $\eta$ (Learning Rate) ÊúÉÂΩ±ÈüøÊØè‰∏ÄÊ¨°Ê≠•Ë™øÁöÑË™øÊï¥„ÄÇ

## Èö®Ê©üÊ¢ØÂ∫¶‰∏ãÈôçÊ≥ï Stochastic Gradient Deacent
- Ê¢ØÂ∫¶‰∏ãÈôçÊ≥ïÂú®ÂØ¶Ë≥™ÁöÑÊáâÁî®‰∏äÔºåÂøÖÈ†àËÄóË≤ªÈùûÂ∏∏Â§ßÁöÑË®àÁÆóÊàêÊú¨„ÄÇ

- Âú®ÁöÑÂØ¶ÂãôÊáâÁî®‰∏äÔºåÂèØÊé°Ë°åÈö®Ê©üÊ¢ØÂ∫¶‰∏ãÈôçÊ≥ï (Stochastic Gradient Deacent, SGD) ‰ΩúÁÇ∫ÂÑ™ÂåñÁÆóÊ≥ï‰æÜÈÄ≤Ë°åÊ¨äÈáçÁöÑÊõ¥Êñ∞„ÄÇ
- SGD ÁöÑÊ¶ÇÂøµÊòØÊØèÊ¨°Ë®àÁÆóÂÉÖÂà©Áî®‰∏ÄÂ∞èÊâπÊ¨°ÁöÑË≥áÊñôÔºåÂè™Ë¶ÅÈö®Ê©üÊäΩÂèñÂá∫Â§†Â§öÊ¨°ÁöÑË≥áÊñôÈªûÔºåÊúÄÁµÇÁöÑÊúüÊúõÂÄºÊúÉËøë‰ººÊñºÊ¢ØÂ∫¶‰∏ãÈôçÊ≥ïÁöÑÁµêÊûú„ÄÇ

## ‰ΩøÁî®Ê¢ØÂ∫¶‰∏ãÈôçÊ≥ïÊ±ÇËß£Á∞°ÂñÆÁ∑öÊÄßËø¥Ê≠∏
- Reference: <a href='https://towardsdatascience.com/linear-regression-using-gradient-descent-97a6c8700931'>Linear Regression using Gradient Descent</a>

### ÂÆöÁæ©‰∏ÄÂÄãÁ∑öÊÄßËø¥Ê≠∏ÂáΩÊï∏ $\Rightarrow f(x) = w_0 + w_1x$

### ÂÆöÁæ©‰∏ÄÂÄãÊêçÂ§±ÂáΩÊï∏ (Loss Function)ÔºåË®àÁÆóÂá∫ÊúÄ‰Ω≥ÊêçÂ§±ÂáΩÊï∏ÁöÑÊ•µÂÄºÔºå
### Ê±ÇËß£Âá∫Êà™Ë∑ù $w_0$ ÂíåÊñúÁéá $w_1$„ÄÇ
### Âú®Ê≠§ÊêçÂ§±ÂáΩÊï∏Êé°Áî® Mean Squared Error (MSE) $\Rightarrow \frac{1}{n}(y_i - \hat y_i)^2$
### $\Rightarrow Loss(w_0, w_1) = \frac{1}{n}\sum_{i}^{n}(y_i - \hat y_i)^2$

### Â∞çÊêçÂ§±ÂáΩÊï∏ Loss ÁöÑÊà™Ë∑ùÂÅèÂæÆÂàÜ
### $\frac{\partial Loss(w_0, w_1)}{\partial w_0}=\frac{1}{n}\frac{\partial \sum_{i}^{n}(y_i - \hat y)^2}{\partial w_0}=0$
### $\Rightarrow \frac{1}{n}\frac{\partial \sum_{i}^{n}(y_i - (w_0 + w_1x_i))^2}{\partial w_0}=0$
### $\Rightarrow \frac{-2}{n}(\sum_{i}^{n}(y_i - w_0 + w_1x_i)=0$
### $\Rightarrow \boxed{\frac{-2}{n}(\sum_{i}^{n}(y_i - \hat y)=0}$

### Â∞çÊêçÂ§±ÂáΩÊï∏ Loss ÁöÑÊñúÁéáÂÅèÂæÆÂàÜ
### $\frac{\partial Loss(w_0, w_1)}{\partial w_1}= \frac{1}{n}\frac{\partial \sum_{i}^{n}(y - \hat y)^2}{\partial w_1} = 0$
### $\Rightarrow  \frac{1}{n} \frac{\partial \sum_{i}^{n}(y_i - (w_0+ w_1x_i))^2}{\partial w_1}=0$
### $\Rightarrow \frac{2}{n}\sum_{i}^{n}(y_i-(w_0+w_1x_i)(-x_i))$
### $\Rightarrow \boxed{\frac{-2}{n}\sum_{i}^{n}x_i(y_i-\hat y)}$

## Á∑öÊÄßËø¥Ê≠∏Ê¢ØÂ∫¶‰∏ãÈôçË®≠ÂÆö

### ÊñπÊ≥ï‰∏Ä

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9, 6)
````

````python
w0 = 0                   # w0 (Êà™Ë∑ù)ÂàùÂßãÂÄº
w1 = 0                   # w1 (ÊñúÁéá)ÂàùÂßãÂÄº
rate = 0.0001              # Â≠∏ÁøíÁéá 0.0001
precision = 0.001        # Áï∂Ê¢ØÂ∫¶‰∏ãÈôçËÆäÂåñÁéáÂ∞èÊñº 0.001 ÊôÇÂÅúÊ≠¢Âü∑Ë°å
max_iters = 1000         # ÊúÄÂ§öÊúâ 1000 Ê¨°Âæ™Áí∞Ë®àÁÆó
step_size = 1            # ÁßªÂãïÁöÑÊîπËÆäÈáè 
iters = 0                # Âæ™Áí∞Ê¨°Êï∏ counter
````

````python
# Preprocessing Input data
#data = pd.read_csv('./data/data.csv', header=None)
data = pd.read_csv('https://bit.ly/3uwhN74', header=None)
X = data.iloc[:, 0]
Y = data.iloc[:, 1]

n = float(len(X)) # X Êï∏Èáè

# ÈÄ≤Ë°åÊ¢ØÂ∫¶‰∏ãÈôçË®àÁÆó
for i in range(max_iters): 
    Y_pred = w0 + w1*X                 # È†êÊ∏¨ÁõÆÂâçÁöÑ y ÂÄº
    #D_w1 = -2 * sum(X * (Y - Y_pred))  # ÊñúÁéá w1 ÁöÑÂ∞éÊï∏
    #D_w0 = -2 * sum(Y - Y_pred)     # Êà™Ë∑ù w0 ÁöÑÂ∞éÊï∏
    D_w1 = (-2/n) * sum(X * (Y - Y_pred))  # ÊñúÁéá w1 ÁöÑÂ∞éÊï∏
    D_w0 = (-2/n) * sum(Y - Y_pred)     # Êà™Ë∑ù w0 ÁöÑÂ∞éÊï∏
    w1 = w1 - rate * D_w1  # Update m
    w0 = w0 - rate * D_w0  # Update c
    #print(f'Êà™Ë∑ù:{round(w0,5)}, ÊñúÁéá:{round(w1,3)}, iters:{i}, D_w0:{round(D_w0,2)}, D_w1:{round(D_w1,2)}')
print(f'Êà™Ë∑ù:{round(w0, 5)}, ÊñúÁéá:{round(w1, 5)}')
````

````python
Y_pred = w0 + w1*X

plt.scatter(X, Y, c='b') 
#plt.plot(X, X * reg.coef_ + reg.intercept_, 'r')
plt.plot(X, Y_pred, 'r')
#plt.plot([min(X), max(X)], [min(Y_pred), max(Y_pred)], color='red')  # regression line
plt.grid(True)
````

### ÊñπÊ≥ï‰∫å

````python
cur_w0 = 0               # w0 (Êà™Ë∑ù)ÂàùÂßãÂÄº
cur_w1 = 0               # w1 (ÊñúÁéá)ÂàùÂßãÂÄº
rate = 0.0001            # Â≠∏ÁøíÁéá 0.0001
precision = 0.00001        # Áï∂Ê¢ØÂ∫¶‰∏ãÈôçËÆäÂåñÁéáÂ∞èÊñº 0.001 ÊôÇÂÅúÊ≠¢Âü∑Ë°å
max_iters = 1000         # ÊúÄÂ§öÊúâ 1000 Ê¨°Âæ™Áí∞Ë®àÁÆó
w0_step_size = 1         # w_0 ÁßªÂãïÁöÑÊîπËÆäÈáè 
w1_step_size = 1         # w_1 ÁßªÂãïÁöÑÊîπËÆäÈáè 
iters = 0                # Âæ™Áí∞Ê¨°Êï∏ counter

# Preprocessing Input data
#data = pd.read_csv('./data/data.csv', header=None)
data = pd.read_csv('https://bit.ly/3uwhN74', header=None)
X = data.iloc[:, 0]
Y = data.iloc[:, 1]
````

````python
while (w0_step_size>precision or w1_step_size>precision) and iters<max_iters:
    Y_pred = cur_w0 + cur_w1*X             # Ë®àÁÆóÁõÆÂâçÁöÑ f ÂÄº
    prev_w0 = cur_w0
    prev_w1 = cur_w1
    
    n = float(len(X)) # X Êï∏Èáè
    
    D_w0 = (-2/n) * sum(Y - Y_pred)        # ÊñúÁéá w0 ÁöÑÂ∞éÊï∏  
    D_w1 = (-2/n) * sum(X * (Y - Y_pred))  # Êà™Ë∑ù w1 ÁöÑÂ∞éÊï∏

    w0_step_size = rate * D_w0             # ÁõÆÂâç w0 step_size
    w1_step_size = rate * D_w1             # ÁõÆÂâç w1 step_size
    
    cur_w0 = cur_w0 - w0_step_size         # ÁõÆÂâç w0 = ÁõÆÂâçw0 - ÁõÆÂâçw0 step_size
    cur_w1 = cur_w1 - w1_step_size         # ÁõÆÂâç w1 = ÁõÆÂâçw1 - ÁõÆÂâçw1 step_size
    
    w0_step_size = abs(cur_w0 - prev_w0)
    w1_step_size = abs(cur_w1 - prev_w1)
    
    iters = iters + 1                  # iteration + 1

print(f'Êà™Ë∑ù:{round(cur_w0, 5)},ÊñúÁéá:{round(cur_w1, 5)},w0_step_size:{round(w0_step_size, 5)}, \
w1_step_size:{round(w1_step_size, 5)},iters:{iters}')
````

````python
cur_x = 0               # x ÂàùÂßãÂÄº
rate = 0.01             # Â≠∏ÁøíÁéá 0.0001
precision = 0.00001     # Áï∂Ê¢ØÂ∫¶‰∏ãÈôçËÆäÂåñÁéáÂ∞èÊñº 0.001 ÊôÇÂÅúÊ≠¢Âü∑Ë°å
previous_step_size = 0.5  
max_iters = 1000        # ÊúÄÂ§öÊúâ 1000 Ê¨°Âæ™Áí∞Ë®àÁÆó

iters = 0               # Âæ™Áí∞Ê¨°Êï∏ counter
df = lambda x: 2*x + 3  # Ê¢ØÂ∫¶‰∏ãÈôçÂáΩÊï∏ 
````

## 4. Ê¢ØÂ∫¶‰∏ãÈôç (Gradient Descent)
- „ÄåÊ¢ØÂ∫¶‰∏ãÈôç„ÄçÊºîÁÆóÊ≥ïÁöÑÁõÆÊ®ôÊòØË¶ÅËóâÁî±Ëø≠‰ª£ (Iterations) Ë®àÁÆóÔºåÂæ™Â∫èÊº∏ÈÄ≤Âú∞ÊêúÂ∞ãÔºåÁõ¥Âà∞ÊâæÂà∞‰∏ÄÂÄãÂáΩÊï∏ÁöÑÂ±ÄÈÉ®Ê•µÂ∞èÂÄº„ÄÇ

````python
plt.rcParams['figure.figsize'] = (7, 5)
x = np.linspace(0, 4, 100)
plt.plot(x, f(x), c='r')

for i in x:
    y = 560 - 200 * x

plt.plot(x, y, c='blue')          # ÂàáÁ∑öÊñúÁéáËÆäÂåñ
plt.scatter(2.8, 560-200*2.8, c='blue', s=75)

plt.text(2.8, 0, '(2.8, 0)')               # Áï´Âá∫Â∫ßÊ®ôÈªû (2.8, 0) 
plt.text(2.8, 1050, '(2.8, 1000)')         # Áï´Âá∫Â∫ßÊ®ôÈªû (2.8, 1000)
plt.scatter(2.8, f(2.8), c='black', s=75)  # Áï´Âá∫Èªû (2.8, 1000)
plt.plot(x, 0*x+1000, c='green')           # Áï´Âá∫Âú®Èªû (0, 1000) ÁöÑÂàáÁ∑ö

plt.ylim(-200, 1200)

plt.grid(True)
````

- Ë®≠ÂÆöÂàùÂßãÂÄº (ÂèØ‰ª•Ë¶ñÊÉÖÊ≥ÅË™øÊï¥):
    - x ÂàùÂßãÂÄº cur_x
    - Â≠∏ÁøíÁéá rate = 0.01
    - x ËÆäÂåñÈáèÈñÄÊ™ªÂÄº precision = 0.001
    - ÊúÄÂ§ßÂæ™Áí∞Ë®àÁÆóÊ¨°Êï∏ max_iters = 50

````python
cur_x = 0                # x ÂàùÂßãÂÄº
rate = 0.0001            # Â≠∏ÁøíÁéá 0.0001
precision = 0.00001      # Áï∂Ê¢ØÂ∫¶‰∏ãÈôçËÆäÂåñÁéáÂ∞èÊñº 0.00001 ÊôÇÂÅúÊ≠¢Âü∑Ë°å
previous_step_size = 1 # 
max_iters = 1000        # ÊúÄÂ§öÊúâ 150 Ê¨°Âæ™Áí∞Ë®àÁÆó

iters = 0              # Âæ™Áí∞Ê¨°Êï∏ counter
df = lambda x: 560 - 200*x   # Ê¢ØÂ∫¶‰∏ãÈôçÂáΩÊï∏ 
````

````python
while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x                    # Â∞áÁõÆÂâç x ÊåáÂÆöÁµ¶Ââç‰∏ÄÊ≠•È©ü
    step_size = rate * df(prev_x)     # step size
    cur_x = cur_x + step_size         # Ê¢ØÂ∫¶‰∏ãÈôç
    previous_step_size = abs(cur_x - prev_x) # Êñ∞ÁöÑ‰ΩçÁΩÆ
    iters = iters + 1                 # iteration + 1
    print("Iteration",iters,"\nX value is",cur_x) 
    
print("The local minimum occurs at", cur_x)
````
