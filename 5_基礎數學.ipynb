## 直線斜率

````python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (7.0, 5.0)
````

````python
w0 = 0.95  # 截距
w1 = 0.64  # 斜率
x = np.linspace(0, 5, 100)
y = w0 + x*w1
plt.plot(x, y, c='r')
plt.scatter(1.5, 1.91, s=75, c='b')
plt.text(1, 2.1, '(x1, y1)', fontsize=12)
plt.scatter(3.5, 3.19, s=75, c='b')
plt.text(3.6, 3.10, '(x2, y2)', fontsize=12)
plt.grid(True)
````

- 在數學上，一條直線的斜率可以經由線上兩點來決定 : 假設有兩點 $(x_1, y_1)$以及$(x_2, y_2)$ 在直線 $L$上，那麼斜率的定義為:
### $Slope = \frac{\Delta y}{\Delta x} = \frac{y_1 - y_2}{x_1 - x_2}$
- 斜率代表直線的陡峭程度以及傾斜方向。

- 若斜率為正，表示只要在這條直線上，往 𝑥 軸的**正向**走，便會越走越高。
- 若斜率為負，表示只要在這條直線上，往 $x$ 軸的**反向**走，便會越走越高。
- 斜率的正負號，決定了這條直線該往哪裡傾斜，但其實也間接地告知，想要往上爬應該往哪個方向走。

## 切線斜率
- 對於任一個曲線 (曲面)上任一點，當找到這一點在某方向上的切線斜率，我們應該可以知道在這一個切點上的傾斜程度及方向。
- 所以，如果要在線上<font color='red'>往上爬</font>，若斜率為正，往 $x$ 軸的正向走；若斜率為負，往 $x$ 軸的反向走。
- 反之，如果要在線上<font color='red'>往下走</font>，若斜率為正，往 $x$ 軸的負向走；若斜率為負，往 $x$ 軸的正向走。

## 合成函數
- 合成函數可將複雜的式子簡化。
- 合成函數的概念是用簡單的函數名稱，取代看似複雜的函數。

### 假設 $y = f(x) = -100(x - 2.8)^2 + 1000$

### 令 $ u = x - 2.8$
### $y = f(x) = -100\boxed{(x - 2.8)}^2 + 1000$
### $\rightarrow f(u) = -100u^2 + 1000$

### 令 $ u = g(x) = x - 2.8$
### 所以 $y = f(u) = -100u^2 + 1000$。
### 將 $f$ 與 $g$ 兩個函數合成在一起等於 $y$
### $\rightarrow \boxed{y = f(g(x))}$

## 連鎖法則 (Chain Rule)

### 當 $y$ 對 $x$ 微分，例如上例的合成函數 $f(g(x))$，必須分別對 $f$ 與 $g$ 函數連續微分:
### $\rightarrow y' = f'(g(x)) \cdot g'(x)$
### $= \boxed{f'(u)\cdot u'}$

### 如果合成函數是由更多的函數一層一層所組織合成，也同樣適用連鎖法則，例如:
### $y = f(g(h(x)))$，將 y 對 x 微分
### $\rightarrow \boxed{y' = f'(g(h(x)) \cdot g'(h(x)) \cdot h'(x)}$

## 使用布萊尼茲記法表示合成函數的連鎖法則微分
### $\frac{dy}{dx} = \frac{dy}{du}\cdot \frac{du}{dx}$
- $y$ 對 $x$微分，等同於「$y$對$u$微分」乘以「$u$對$x$微分」
### 因為$ y = f(u) = f(g(x)) $
### 所以，$ \boxed{\frac{dy}{dx} = \frac{dy}{du}\cdot \frac{du}{dx}} \equiv \boxed{f'(u)\cdot u'}$

## 範例

## $\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx}$

### 而 $\frac{dy}{du} = \frac{d}{du}f(u) = f'(u) = -200\cdot u = -200u$

### 而 $\frac{du}{dx} = \frac{d}{dx}(x - 2.8) = 1$

### 所以，$\frac{dy}{dx} = \frac{dy}{du} \cdot \frac{du}{dx} = -200u \cdot 1 = -200(x - 2.8)$

## 偏導數
- 如果題目是有關高維度 (3 度空間以上) 的曲面，該如何描述在這個曲面上某一點的陡峭程度?
- 利用偏導函數可求出在高維度曲面中，某一點在各方向上的切線斜率(偏導數)。

- 假設 $f(x_1, x_2, \cdots, x_n)$ 為 $\mathbb{R}^n$ 空間中一曲面方程式，求出 $x_i$ 在各方向上的偏導函數:
### $$\frac{\partial f(x_1, x_2, \cdots, x_n)}{\partial x_i}$$
- 其中，某一點 $x$ 在 $x_i$ 方向的切線斜率等同於此 $x$ 點在 $f(x_1, x_2, \cdots, x_n)$ 方向的傾斜程度。
- 而偏導數在此點的正號與負號，表示該點在 $f(x_1, x_2, \cdots, x_n)$ 各軸向的上升或下降方向。
- 例如，在 3 度空間 $f(x, y, z)$ 的曲面，當 $\frac{\partial f(x, y, z)}{\partial x}$ 在 $x$ 方向為正，表示當往 $x$ 軸的正向走，斜率上升；反之斜率下降。

## 多變數函數
- 有兩個或兩個以上變數的函數，稱為多變數函數。
- 偏微分是在多變數函數中，僅針對某一個變數進行微分，與該變數無關的其它變數會被視為常數處理。

### 函數 $z = f(x, y) = x + y$，其中 $x$ 與 $y$ 互相獨立，不受影響。
### 令 z 對 x 微分時，y 可視為與 x 無關的常數，其偏微分的表示如下:
### $\frac{\partial z}{\partial x} = \frac{\partial}{\partial x}(x + y) = \frac{\partial}{\partial x}(x) + \frac{\partial}{\partial x}(y) = 1 + 0 = 1$

### 再看一個範例
### $f(x, y) = e^{2y}sin(x)$
###  $\Rightarrow$ 假如函數 $f$ 僅針對變數 $x$ 偏微分, 則 y 變數視為常數 
### $\therefore \frac{\partial f}{\partial x} (x, y) = \boxed{e^{2y}cos(x)}$
### $\Rightarrow$ 假如函數 $f$ 僅針對變數 $y$ 偏微分, 則 x 變數視為常數 
### $\therefore \frac{\partial f}{\partial y} (x, y) = \boxed{2e^{2y}sin(x)}$

## 梯度
- 如果把上述每一個方向的偏導函數收集起來，稱為梯度。
### $$\triangledown f(x_1, x_2, \cdots, x_n) = (\frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \cdots, \frac{\partial f}{\partial x_n})$$
- **梯度，可以表現此高為曲面在某一點上，對於每一個方向的描述。不僅個別去看此點在每一個軸上的傾斜表現，梯度包含了所有的資訊在其中。**
- 所以，假若我們想要找到 $f(x_1, x_2, \cdots, x_n)$ 的極值，雷同於在平面空間對一條曲線方程式的切線斜率，最簡單的方式是令 $\Delta f(x_1, x_2, \cdots, x_n) = 0$。

## 再以前述 $f(x,y)=e^{2y}sin(x)$函數為例
### 函數 $f$ 僅針對變數 $x$ 偏微分，則 $y$ 變數視為常數 $\frac{\partial f}{\partial x} (x, y) = \boxed{e^{2y}cos(x)}$
### 函數 $f$ 僅針對變數 $y$ 偏微分，則 $x$ 變數視為常數$\frac{\partial f}{\partial y} (x, y) = \boxed{2e^{2y}sin(x)}$

### 那麼，函數$f(x,y)=e^{2y}sin(x)$的梯度下降$\Delta f(x,y)$表達如下:
### $\triangledown f(x,y) = (\frac{\partial f}{\partial x}(x, y), \frac{\partial f}{\partial y}(x, y)) = \boxed{(e^{2y}cos(x), 2e^{2y}sin(x))}$

## 梯度下降 (Gradient Descent)
- 「梯度下降」演算法 (Gradient Descent Algorithm)是大多數機器學習( machime Learning ）和深度學習 (Deep Learning) 的核心內容。
- 「梯度下降」演算法的目標是要直到找到一個函數的最佳解，概念是根據隨機選擇函數的初始點，藉由迭代 (Iterations) 計算，，循序漸進地搜尋，不斷調整步伐以及方向，試圖找出最適合的解。
- 在一個損失函數 (Loss Function) 上，隨機找一個點，進行迭代，移動到下一個點，直到不再移動或無法移動為止。

### $$w^{(t+1)} = w^t + \eta \cdot (-\triangledown L(w^t)) = \boxed{w^t - \eta \cdot \triangledown L(w^t)}$$

### 說明 : 下降梯度新的點 ($w^{t+1}$) 的位置是依靠舊的點 ($W^t$) 梯度正負號的反方向，以及學習率大小 ($\eta$) 計算得知。
- $w^t$: 目前的位置
- $w^{(t+1)}$: 新的位置
- $t$: iteration 迭代, $t\leftarrow$ 是從第 0 次開始<br>
- $\eta$: 學習率 (Learning Rate)
- $\triangledown L$: 梯度下降函數的變化
- $-\triangledown L(w^t)$: 設定負號表示損失(loss)愈小愈好，確定梯度是往最小方向移動。

### 梯度下降法演算法 : 
$1: Guess\; w^{(0)}, set \; t\leftarrow \, 0 $<br>
$2: \textbf{while} \,||\triangledown f(w^{(t)}) || \geq \epsilon \; \textbf{do}  $<br>
$3: \;\;\;\; w^{(k+1)} = |w^{(k)} - \eta \triangledown f(w^{(k)})|  $<br>
$4: \;\;\;\; t \leftarrow t + 1 $<br>
$5: \textbf{end while} $<br>
$6: \textbf{return} \;\; w^{(k)}$ <p>

### 梯度下降法演算法說明:
$w^{(0)}: w \;的初始設定$ <br>
$t: iteration$;  $t\leftarrow \; 0$ 是從第 0 次開始<br>
$\triangledown f(w^{(t)}):\; $ 表示梯度下降函數在第 t 次的變化率 <br>
$||\triangledown f(w^{(t)}) || \geq \epsilon:\;$ 表示是第 t 個 iteration 時，若大於 $\epsilon$ 變化值，則進入 iteration <br>
$ w^{(k+1)} = w^{(k)} - \eta \triangledown f(w^{(t)}): \;$ <br>
$ \;\;\;\; w^{(t+1)}\; 是新的 w$ <br>
$ \;\;\;\; w^{(t)}\; 是舊的 x, $ <br>
$ \;\;\;\; \eta \,是\;learning \;rate$ <br>
$\therefore 新的 w^{(t+1)} \; 等於舊的 \;w^k\;減去\;learning rate \;\eta \;乘以梯度下降\; \triangledown f(w^{(t)})$ <br>
$t \leftarrow t + 1: \;$ 再進入下一個循環<br>

## 梯度下降 (Gradient Descent) 特性
- 必須要確定 Loss function 本身可微分 (differentiable)。
- 選擇不同的起始點，Gradient Descent過程會非常不一樣。
- 學習率 $\eta$ (Learning Rate) 會影響每一次步調的調整。

- 每一次的梯度計算都必須要將所有資料納入計算。

## 隨機梯度下降法 Stochastic Gradient Deacent
- 梯度下降法在實質的應用上，必須耗費非常大的計算成本。
- 因為每一次的更新，都必須要計算一次梯度，而每一次計算梯度都必須將計算出所有資料才能得到梯度。
- 在的實務應用上，可採行隨機梯度下降法 (Stochastic Gradient Deacent, SGD) 作為優化算法來進行權重的更新。
- SGD 的概念是每次計算僅利用一小批次的資料，只要隨機抽取出夠多次的資料點，最終的期望值會近似於梯度下降法的結果。
- 使用 SGD 取代 GD 的確會讓整個計算成本大幅下降，但由於隨機性的關係，SGD 也會相對不穩定。

## 使用梯度下降法求解簡單線性迴歸
- Reference: <a href='https://towardsdatascience.com/linear-regression-using-gradient-descent-97a6c8700931'>Linear Regression using Gradient Descent</a>

### 定義一個線性迴歸函數 $\Rightarrow f(x) = w_0 + w_1x$

### 定義一個損失函數 (Loss Function)，計算出最佳損失函數的極值，
### 求解出截距 $w_0$ 和斜率 $w_1$。
### 在此損失函數採用 Mean Squared Error (MSE) $\Rightarrow \frac{1}{n}(y_i - \hat y_i)^2$
### $\Rightarrow Loss(w_0, w_1) = \frac{1}{n}\sum_{i}^{n}(y_i - \hat y_i)^2$

### 對損失函數 Loss 的截距偏微分
### $\frac{\partial Loss(w_0, w_1)}{\partial w_0}=\frac{1}{n}\frac{\partial \sum_{i}^{n}(y_i - \hat y)^2}{\partial w_0}=0$
### $\Rightarrow \frac{1}{n}\frac{\partial \sum_{i}^{n}(y_i - (w_0 + w_1x_i))^2}{\partial w_0}=0$
### $\Rightarrow \frac{-2}{n}(\sum_{i}^{n}(y_i - w_0 + w_1x_i)=0$
### $\Rightarrow \boxed{\frac{-2}{n}(\sum_{i}^{n}(y_i - \hat y)=0}$

### 對損失函數 Loss 的斜率偏微分
### $\frac{\partial Loss(w_0, w_1)}{\partial w_1}= \frac{1}{n}\frac{\partial \sum_{i}^{n}(y - \hat y)^2}{\partial w_1} = 0$
### $\Rightarrow  \frac{1}{n} \frac{\partial \sum_{i}^{n}(y_i - (w_0+ w_1x_i))^2}{\partial w_1}=0$
### $\Rightarrow \frac{2}{n}\sum_{i}^{n}(y_i-(w_0+w_1x_i)(-x_i))$
### $\Rightarrow \boxed{\frac{-2}{n}\sum_{i}^{n}x_i(y_i-\hat y)}$

## 線性迴歸梯度下降設定

### 方法一

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (9, 6)
````

````python
w0 = 0                   # w0 (截距)初始值
w1 = 0                   # w1 (斜率)初始值
rate = 0.0001              # 學習率 0.0001
precision = 0.001        # 當梯度下降變化率小於 0.001 時停止執行
max_iters = 1000         # 最多有 1000 次循環計算
step_size = 1            # 移動的改變量 
iters = 0                # 循環次數 counter
````

````python
# Preprocessing Input data
#data = pd.read_csv('./data/data.csv', header=None)
data = pd.read_csv('https://bit.ly/3uwhN74', header=None)
X = data.iloc[:, 0]
Y = data.iloc[:, 1]

n = float(len(X)) # X 數量

# 進行梯度下降計算
for i in range(max_iters): 
    Y_pred = w0 + w1*X                 # 預測目前的 y 值
    #D_w1 = -2 * sum(X * (Y - Y_pred))  # 斜率 w1 的導數
    #D_w0 = -2 * sum(Y - Y_pred)     # 截距 w0 的導數
    D_w1 = (-2/n) * sum(X * (Y - Y_pred))  # 斜率 w1 的導數
    D_w0 = (-2/n) * sum(Y - Y_pred)     # 截距 w0 的導數
    w1 = w1 - rate * D_w1  # Update m
    w0 = w0 - rate * D_w0  # Update c
    #print(f'截距:{round(w0,5)}, 斜率:{round(w1,3)}, iters:{i}, D_w0:{round(D_w0,2)}, D_w1:{round(D_w1,2)}')
print(f'截距:{round(w0, 5)}, 斜率:{round(w1, 5)}')
````

## 以梯度下降法，求出 $f(x) = x^2 + 3x -5$函數的最小值

## 導函數 $f'(x) = 2x + 3$
- 當然，從導函數可以知道當 $x=-1.5$時導函數為 0 有最小值。
- 試著使用梯度下降演算法求出最小值。

````python
import numpy as np
import matplotlib.pyplot as plt
````

````python
cur_x = 0               # x 初始值
rate = 0.01             # 學習率 0.0001
precision = 0.00001     # 當梯度下降變化率小於 0.001 時停止執行
previous_step_size = 0.5  
max_iters = 1000        # 最多有 1000 次循環計算

iters = 0               # 循環次數 counter
df = lambda x: 2*x + 3  # 梯度下降函數 
````

````python
xList = []
while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x                    # 將目前 x 指定給前一步驟
    step_size = rate * df(prev_x)     # step size
    cur_x = cur_x - step_size         # 梯度下降
    xList.append(cur_x)
    previous_step_size = abs(cur_x - prev_x) # 新的位置
    iters = iters + 1                 # iteration + 1
    #print("Iteration",iters,"X: ",cur_x) 
    
print("The local minimum occurs at", round(cur_x,3))
````

````python
# 畫出梯度下降 cur_x 的變化
plt.plot(xList)
plt.xlabel('Running Times', fontsize=15)
plt.ylabel('f\'(x)=2x+3', fontsize=15)
plt.grid(True)
````

## 4. 梯度下降 (Gradient Descent)
- 「梯度下降」演算法的目標是要藉由迭代 (Iterations) 計算，循序漸進地搜尋，直到找到一個函數的局部極小值。

````python
plt.rcParams['figure.figsize'] = (7, 5)
x = np.linspace(0, 4, 100)
plt.plot(x, f(x), c='r')

for i in x:
    y = 560 - 200 * x

plt.plot(x, y, c='blue')          # 切線斜率變化
plt.scatter(2.8, 560-200*2.8, c='blue', s=75)

plt.text(2.8, 0, '(2.8, 0)')               # 畫出座標點 (2.8, 0) 
plt.text(2.8, 1050, '(2.8, 1000)')         # 畫出座標點 (2.8, 1000)
plt.scatter(2.8, f(2.8), c='black', s=75)  # 畫出點 (2.8, 1000)
plt.plot(x, 0*x+1000, c='green')           # 畫出在點 (0, 1000) 的切線

plt.ylim(-200, 1200)

plt.grid(True)
````

- 設定初始值 (可以視情況調整):
    - x 初始值 cur_x
    - 學習率 rate = 0.01
    - x 變化量門檻值 precision = 0.001
    - 最大循環計算次數 max_iters = 50

````python
cur_x = 0                # x 初始值
rate = 0.0001            # 學習率 0.0001
precision = 0.00001      # 當梯度下降變化率小於 0.00001 時停止執行
previous_step_size = 1 # 
max_iters = 1000        # 最多有 150 次循環計算

iters = 0              # 循環次數 counter
df = lambda x: 560 - 200*x   # 梯度下降函數 
````

````python
while previous_step_size > precision and iters < max_iters:
    prev_x = cur_x                    # 將目前 x 指定給前一步驟
    step_size = rate * df(prev_x)     # step size
    cur_x = cur_x + step_size         # 梯度下降
    previous_step_size = abs(cur_x - prev_x) # 新的位置
    iters = iters + 1                 # iteration + 1
    print("Iteration",iters,"\nX value is",cur_x) 
    
print("The local minimum occurs at", cur_x)
````
